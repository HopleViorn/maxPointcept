```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Define the target function y(x)
def target_function(x):
    return 1.2 * np.sin(np.pi * x) * np.cos(2.4 * np.pi * x)

# Generate training and test data
def generate_data(step_size):
    x = np.arange(-2, 2 + step_size, step_size)
    y = target_function(x)
    return x, y

def generate_data_2(step_size):
    x = np.arange(-4, 4 + step_size, step_size)
    y = target_function(x)
    return x, y

# Prepare the data
x_train, y_train = generate_data(0.05)  # Training set
x_test, y_test = generate_data_2(0.01)    # Test set

# Convert data to PyTorch tensors
x_train = torch.tensor(x_train, dtype=torch.float32).reshape(-1, 1)
y_train = torch.tensor(y_train, dtype=torch.float32).reshape(-1, 1)
x_test = torch.tensor(x_test, dtype=torch.float32).reshape(-1, 1)
y_test = torch.tensor(y_test, dtype=torch.float32).reshape(-1, 1)


# Convert data to PyTorch tensors
x_train = torch.tensor(x_train, dtype=torch.float32).reshape(-1, 1)
y_train = torch.tensor(y_train, dtype=torch.float32).reshape(-1, 1)
x_test = torch.tensor(x_test, dtype=torch.float32).reshape(-1, 1)
y_test = torch.tensor(y_test, dtype=torch.float32).reshape(-1, 1)

# Define the MLP model with Tanh activation
class MLP(nn.Module):
    def __init__(self, n_neurons):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(1, n_neurons)
        self.fc2 = nn.Linear(n_neurons, 1)
        self.tanh = nn.Tanh()  # Use Tanh activation function
    
    def forward(self, x):
        x = self.tanh(self.fc1(x))
        x = self.fc2(x)
        return x

# Training function with Adam optimizer (simulating `trainbr`)
def train_mlp_adam(model, x_train, y_train, n_epochs=10000, lr=0.001, weight_decay=0.0001):
    # Define loss and optimizer with L2 regularization (weight_decay)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)
    
    # Training loop
    loss_history = []
    for epoch in range(n_epochs):
        model.train()
        optimizer.zero_grad()
        y_pred = model(x_train)
        loss = criterion(y_pred, y_train)
        loss.backward()
        optimizer.step()
        loss_history.append(loss.item())
        
        # Print progress every 1000 epochs
        if (epoch+1) % 1000 == 0:
            print(f'Epoch [{epoch+1}/{n_epochs}], Loss: {loss.item():.6f}')
    
    return loss_history

# Training function with LBFGS optimizer (simulating `trainlm`)
def train_mlp_lbfgs(model, x_train, y_train, n_epochs=10000):
    # Define loss and optimizer (using LBFGS)
    criterion = nn.MSELoss()
    optimizer = optim.LBFGS(model.parameters(), max_iter=100)
    
    # Training loop
    loss_history = []
    def closure():
        optimizer.zero_grad()
        y_pred = model(x_train)
        loss = criterion(y_pred, y_train)
        loss.backward()
        return loss
    
    for epoch in range(n_epochs):
        model.train()
        optimizer.step(closure)
        with torch.no_grad():
            loss = criterion(model(x_train), y_train)
        loss_history.append(loss.item())
        
        # Print progress every 1000 epochs
        if (epoch+1) % 1000 == 0:
            print(f'Epoch [{epoch+1}/{n_epochs}], Loss: {loss.item():.6f}')
    
    return loss_history

# Function to plot predictions
def plot_predictions(model, x_test, y_test, x_train, y_train, title=""):
    model.eval()
    with torch.no_grad():
        y_pred_test = model(x_test).numpy()
        y_pred_train = model(x_train).numpy()
    
    # Plot the true function and predictions
    plt.figure(figsize=(10, 6))
    plt.ylim((-1.5,1.5))
    plt.plot(x_test.numpy(), y_test.numpy(), 'b-', label='True Function')
    plt.plot(x_test.numpy(), y_pred_test, 'r--', label='Predicted Function (Test)')
    plt.scatter(x_train.numpy(), y_train.numpy(), label='Training Data', c='green')
    plt.title(title)
    plt.legend()
    plt.show()

# Experiment with different numbers of hidden neurons
n_neurons_list = [1, 2, 5, 10, 20, 50, 100]

# Training with Adam optimizer (trainbr simulation)
for n_neurons in n_neurons_list:
    print(f'\nTraining MLP with {n_neurons} hidden neurons using Adam (trainbr)...')
    model = MLP(n_neurons=n_neurons)
    train_mlp_adam(model, x_train, y_train, n_epochs=10000, lr=0.001, weight_decay=0.0001)
    plot_predictions(model, x_test, y_test, x_train, y_train, title=f'MLP with {n_neurons} Hidden Neurons (sequential)')

# Training with LBFGS optimizer (trainlm simulation)
for n_neurons in n_neurons_list:
    print(f'\nTraining MLP with {n_neurons} hidden neurons using LBFGS (trainlm)...')
    model = MLP(n_neurons=n_neurons)
    train_mlp_lbfgs(model, x_train, y_train, n_epochs=10000)
    plot_predictions(model, x_test, y_test, x_train, y_train, title=f'MLP with {n_neurons} Hidden Neurons (trainlm)')
```
