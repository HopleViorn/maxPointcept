```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Define the target function y(x)
def target_function(x):
    return 1.2 * np.sin(np.pi * x) * np.cos(2.4 * np.pi * x)

def compute_mlp_output(model, x_values):
    model.eval()  # Set the model to evaluation mode
    x_tensor = torch.tensor(x_values, dtype=torch.float32).reshape(-1, 1)
    with torch.no_grad():
        outputs = model(x_tensor)
    return outputs.numpy()

# Generate training and test data
def generate_data(step_size):
    x = np.arange(-2, 2 + step_size, step_size)
    y = target_function(x)
    return x, y

def generate_data_2(step_size):
    x = np.arange(-4, 4 + step_size, step_size)
    y = target_function(x)
    return x, y

x_train, y_train = generate_data(0.05)  # Training set
x_test, y_test = generate_data_2(0.01)    # Test set

x_train = torch.tensor(x_train, dtype=torch.float32).reshape(-1, 1)
y_train = torch.tensor(y_train, dtype=torch.float32).reshape(-1, 1)
x_test = torch.tensor(x_test, dtype=torch.float32).reshape(-1, 1)
y_test = torch.tensor(y_test, dtype=torch.float32).reshape(-1, 1)

class MLP(nn.Module):
    def __init__(self, n_neurons):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(1, n_neurons)
        self.fc2 = nn.Linear(n_neurons, 1)
        self.relu = nn.Sigmoid()  # You can also use Tanh or Sigmoid for activation
    
    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x
    
def train_mlp(model, x_train, y_train, n_epochs=1000, lr=0.01 ,weight_decay = 0.0001):
    criterion = nn.MSELoss()
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay = 0.0001)
    
    loss_history = []
    for epoch in range(n_epochs):
        model.train()
        optimizer.zero_grad()
        y_pred = model(x_train)
        loss = criterion(y_pred, y_train)
        loss.backward()
        optimizer.step()
        loss_history.append(loss.item())
        
        if (epoch+1) % 100 == 0:
            print(f'Epoch [{epoch+1}/{n_epochs}], Loss: {loss.item():.6f}')
    
    return loss_history

def plot_predictions(model, x_test, y_test, x_train, y_train, title=""):
    model.eval()
    with torch.no_grad():
        y_pred_test = model(x_test).numpy()
        y_pred_train = model(x_train).numpy()
    
    plt.xlim((-3.5,3.5))
    plt.plot(x_test.numpy(), y_test.numpy(), 'b-', label='True Function')
    plt.plot(x_test.numpy(), y_pred_test, 'r--', label='Predicted Function (Test)')
    plt.scatter(x_train.numpy(), y_train.numpy(), label='Training Data', c='green')
    plt.title(title)
    plt.legend()
    plt.show()

n_neurons_list = [1, 2, 3, 5, 10, 20, 50, 100]

plt.figure(figsize=(10, 6))
for i, n_neurons in enumerate(n_neurons_list):
    print(f'\nTraining MLP with {n_neurons} hidden neurons...')
    model = MLP(n_neurons=n_neurons)
    train_mlp(model, x_train, y_train, n_epochs=10000, lr=0.01)
    print(f'prediction at -3.0 :{compute_mlp_output(model,-3.0)}, ground truth:{target_function(-3.0):4f}')
    print(f'prediction at 3.0 :{compute_mlp_output(model,3.0)}, ground truth:{target_function(3.0):4f}')
    plot_predictions(model, x_test, y_test, x_train, y_train, title=f'MLP with {n_neurons} Hidden Neurons')
```
